## Crear el dataframe, primer vistazo y análisis de duplicados

## Pruebas de hipótesis

+ Para definir H0 y H1, me baso en las preguntas que se plantean:
  + La duración promedio de los viajes desde el Loop hasta el Aeropuerto Internacional O'Hare cambia los sábados lluviosos (H0)
  + La duración promedio de los viajes desde el Loop hasta el Aeropuerto Internacional O'Hare ES IGUAL los sábados lluviosos (H1)
+ Dado que sólo se pide encontrar si los promedios son diferentes, es decir, no se especifica si es mayor o menor que, ejecuto una prueba de hipótesis de dos colas con la prueba t de Student para muestras independientes.

```python
clean_loop_ohare_trips.head(10)
#Convertir la columna start_ts a tipo datetime
clean_loop_ohare_trips['start_ts'] = pd.to_datetime(clean_loop_ohare_trips['start_ts'])
clean_loop_ohare_trips.info()
#Extraer el nombre de día de la semana de la columna start_ts y crear la columna weekday
clean_loop_ohare_trips['weekday'] = clean_loop_ohare_trips['start_ts'].dt.day_name()
# df['day_of_week'] = df['date'].dt.day_name()
display(clean_loop_ohare_trips)
print(clean_loop_ohare_trips['weekday'].unique()) #Todos los registros corresponden a días sábados
print(clean_loop_ohare_trips['weather_conditions'].unique()) #Hay dos posibles condiciones metereológicas, Good y Bad
```

```python
#Filtro y creo dos dataframes para almacenar los viajes en días de condiciones metereológicas buenas y malas
good_weather = clean_loop_ohare_trips.query('weather_conditions=="Good"')
display(good_weather)
bad_weather = clean_loop_ohare_trips.query('weather_conditions=="Bad"')
display(bad_weather)
```

## Prueba de Levene para varianzas

```python
#Prueba de Levene para las varianzas
#H0: Las varianzas para ambos dataframes son IGUALES
#H1: Las varianzas para ambos dataframes son DIFERENTES
good_weather_var=np.var(good_weather['duration_seconds'])
bad_weather_var=np.var(bad_weather['duration_seconds'])
print(f'Good weather variance: {good_weather_var}')
print(f'Bad weather variance: {bad_weather_var}')
alpha=0.05
levene_result = st.levene(good_weather['duration_seconds'],bad_weather['duration_seconds'])
print(levene_result)

pvalue=levene_result.pvalue

if pvalue < alpha:
    print("Se rechaza la hipótesis nula. Las varianzas para ambas condiciones SON DIFERENTES")
else:
    print("No se puede rechazar la hipótesis nula. Las varianzas son significativamente iguales.")
```

## Prueba de hipótesis t de Student

```python
# Prueba las hipótesis
alpha = 0.05
test = st.ttest_ind(good_weather['duration_seconds'],bad_weather['duration_seconds'],equal_var=True)
pvalue = test.pvalue 
print(pvalue)
if pvalue < alpha:
    print("Se rechaza la hipótesis nula. Las duraciones promedio de los viajes en ambas condiciones SON DIFERENTES.")
else:
    print("No se puede rechazar la hipótesis nula.")
```
