# TOP 6 PLATAFORMAS CON MÁS VENTAS

```python
#Creo una lista con las plataformas con más ventas
top_sales_platforms = ['PS2','X360','PS3','Wii','DS','PS']

#Creo un nuevo dataframe que incluya sólo los juegos cuya plataforma se incluya en la lista anterior
top_platforms = clean_games[clean_games['platform'].isin(top_sales_platforms)]

#Agrupo lo juegos por plataforma y año de lanzamiento
pivot_top_platforms = top_platforms.pivot_table(index=['platform','year_of_release'], values='total_sales', aggfunc='sum').reset_index()
# print(pivot_top_platforms)

#Agrupo nuevamente para cada plataforma, filtrando los juegos que no tenían año de lanzamiento
pivot_DS = pivot_top_platforms.query('platform=="DS" and year_of_release != 9999 and year_of_release > 2010')[['year_of_release','total_sales']].set_index('year_of_release')
pivot_PS2 = pivot_top_platforms.query('platform=="PS2" and year_of_release != 9999 and year_of_release > 2010')[['year_of_release','total_sales']].set_index('year_of_release')
pivot_X360 = pivot_top_platforms.query('platform=="X360" and year_of_release != 9999 and year_of_release > 2010')[['year_of_release','total_sales']].set_index('year_of_release')
pivot_Wii = pivot_top_platforms.query('platform=="Wii" and year_of_release != 9999 and year_of_release > 2010')[['year_of_release','total_sales']].set_index('year_of_release')
# pivot_PS = pivot_top_platforms.query('platform=="PS" and year_of_release != 9999 and year_of_release > 2010')[['year_of_release','total_sales']].set_index('year_of_release')
pivot_PS3 = pivot_top_platforms.query('platform=="PS3" and year_of_release != 9999 and year_of_release > 2010')[['year_of_release','total_sales']].set_index('year_of_release')
display(pivot_DS)
#Creo los gráficos de baras para cada plataforma
fig, axes = plt.subplots(2,3, figsize=(25,10))
pivot_DS.plot(kind='bar',title='DS', xlabel='Year of release', ylabel='Total sales in million USD', ax= axes[0,0])
pivot_PS2.plot(kind='bar',title='PS2', xlabel='Year of release', ylabel='Total sales in million USD', ax= axes[0,1])
pivot_X360.plot(kind='bar',title='X360', xlabel='Year of release', ylabel='Total sales in million USD', ax= axes[0,2])
pivot_Wii.plot(kind='bar',title='Wii', xlabel='Year of release', ylabel='Total sales in million USD', ax= axes[1,0])
# pivot_PS.plot(kind='bar',title='PS', xlabel='Year of release', ylabel='Total sales in million USD', ax= axes[1,1])
pivot_PS3.plot(kind='bar',title='PS3', xlabel='Year of release', ylabel='Total sales in million USD', ax= axes[1,2])
plt.tight_layout()
plt.show()
```

+ Las plataformas X360, PS y PS3 muestran una distribución normal de ventas, mientras que DS, PS2 y Wii muestran una asimetría positiva.

## ¿CÓMO AFECTAN LAS RESEÑAS A LAS VENTAS?

```python
#Filtro los juegos por aquellos que sí tenían un score de la crítica y de los usuarios, sólo para la plataforma PS2
reviews=clean_games.query('critic_score<1000 and user_score<1000 and platform=="PS2"')
# print(reviews['critic_score'].unique())
display(reviews)

#Creo las gráficas de distribución para el score de los críticos y de los usuarios
fig, axes = plt.subplots(1,2, figsize=(15,5))
reviews.plot(kind='scatter', y='total_sales', x='critic_score', ax=axes[0])
reviews.plot(kind='scatter', y='total_sales', x='user_score', ax=axes[1])
plt.tight_layout()
plt.show()

#Calculo los valores de correlación
print(reviews['critic_score'].corr(reviews['total_sales']))
print(reviews['user_score'].corr(reviews['total_sales']))
print(reviews[['critic_score','user_score','total_sales']].corr())
```

### Comparación de ventas totales de juegos en 2 de las plataformas más populares, PS2 y X360. Sólo para juegos con un score de la crítica mayor a 70 puntos y excluyendo todos los juegos que no tenían un score asignado (critic_score = 1000)

No parece haber alguna relación entre la plataforma y las ventas totales si comparamos estas dos plataformas. Algunos juegos tienen sólo la mitad de ventas en una plataforma que en la otra, aún tratándose del mismo juego.
En este análisis, el score de la crítica sólo se usó para filtrar los juegos y no presentar la lista completa.

```python
#Filtro los juegos por plataforma para PS2 y X360, tomando sólo los que tienen un score de la crítica payor a 70
PS2_games = clean_games[(clean_games['platform']=='PS2') & (clean_games['critic_score']>70) & (clean_games['critic_score']<1000)][['name','total_sales']]
X360_games = clean_games[(clean_games['platform']=='X360') & (clean_games['critic_score']>70) & (clean_games['critic_score']<1000)][['name','total_sales']]

#Mezclo las dos tablas para poder crear la gráfica de barras que muestre las ventas totales de juegos que se lanzaron en ambas plataformas
PS2_X360 = PS2_games.merge(X360_games, on='name', how='inner')
# display(PS2_X360)
PS2_X360.plot(kind='bar',figsize=[20,5],title='Games launched for PS2 and X360',x='name',xlabel='Game name',ylabel='Total sales')
plt.legend(['PS2','X360'])
plt.show()
```

# Pruebas de hipótesis

+ Para definir H0 y H1, me baso en las preguntas que se plantean:
  + Las calificaciones promedio de los usuarios para las plataformas Xbox One y PC SON LAS MISMAS (H0 del planteamiento 1)
  + Las calificacione spromedio de los usarios para los géneros de acción y deportes SON DIFERENTES (H1 del planteamiento 2)
+ Las hipótesis alternativa y nula faltantes en ambos planteamientos, las creo como un planteamiento contrario de las que ya están definidas.
+ Dado que sólo se pide encontrar si los promedios son diferentes, es decir, no se especifica si es mayor o menor que, ejecuto una prueba de hipótesis de dos colas con la prueba t de Student para muestras independientes.

## Calificaciones promedio de usuarios para plataformas Xbox One y PC

Hipótesis nula: Las calificaciones promedio de los usuarios para las plataformas Xbox One y PC son las mismas.

Hipótesis alternativa: Las calificaciones promedio de los usuarios para las plataformas Xbox One y PC SON DIFERENTES.

```python
#Filtro los juegos que ya tenían un score de usuario (distinto de 1000) y que pertenencen a las plataformas XOne y PC
xone_user_rating = clean_games.query('user_score<1000 and platform=="XOne"')
display(xone_user_rating)
pc_user_rating = clean_games.query('user_score<1000 and platform=="PC"')
display(pc_user_rating)
```

```python
#Prueba de Levene para las varianzas
#H0: Las varianzas para ambas plataformas son iguales
#H1: Las varianzas para ambas plataformas son diferentes
xone_var=np.var(xone_user_rating['user_score'])
pc_var=np.var(pc_user_rating['user_score'])
print(f'Xone variance: {xone_var}')
print(f'PC variance: {pc_var}')
alpha=0.05
levene_result = st.levene(xone_user_rating['user_score'],pc_user_rating['user_score'])
print(levene_result)

pvalue=levene_result.pvalue

if pvalue < alpha:
    print("Se rechaza la hipótesis nula. Las varianzas para ambas plataformas SON DIFERENTES")
else:
    print("No se puede rechazar la hipótesis nula. Las varianzas son significativamente iguales.")
```

```python
# Prueba las hipótesis
alpha = 0.05
test = st.ttest_ind(xone_user_rating['user_score'],pc_user_rating['user_score'],equal_var=True)
pvalue = test.pvalue 
print(pvalue)
if pvalue < alpha:
    print("Se rechaza la hipótesis nula. Las calificaciones promedio de los usuarios para ambas plataformas SON DIFERENTES.")
else:
    print("No se puede rechazar la hipótesis nula.")
```

## Calificaciones promedio de usuarios para géneros de acción y deportes

Hipótesis nula: Las calificaciones promedio de los usuarios para los géneros de acción y deportes son las mismas.

Hipótesis alternativa: Las calificaciones promedio de los usuarios para los géneros de acción y deportes son diferentes.

```python
#Filtro los juegos que ya tenían un score de usuario (distinto de 1000) y que pertenencen a lor géneros de acción y deportes
action_user_rating = clean_games.query('user_score<1000 and genre=="Action"')
display(action_user_rating)
sports_user_rating = clean_games.query('user_score<1000 and genre=="Sports"')
display(sports_user_rating)
```

```python
#Prueba de Levene para las varianzas
#H0: Las varianzas para ambos géneros son iguales
#H1: Las varianzas para ambos géneros son diferentes
action_var=np.var(action_user_rating['user_score'])
sports_var=np.var(sports_user_rating['user_score'])
print(f'Action variance: {action_var}')
print(f'Sports variance: {sports_var}')
alpha=0.05
levene_result = st.levene(action_user_rating['user_score'],sports_user_rating['user_score'])
print(levene_result)

pvalue=levene_result.pvalue

if pvalue < alpha:
    print("Se rechaza la hipótesis nula. Las varianzas para ambos géneros SON DIFERENTES")
else:
    print("No se puede rechazar la hipótesis nula. Las varianzas son significativamente iguales.")
```

```python
# Prueba las hipótesis
alpha = 0.01
test = st.ttest_ind(action_user_rating['user_score'],sports_user_rating['user_score'],equal_var=False)
pvalue = test.pvalue 
print(pvalue)
if pvalue < alpha:
    print("Se rechaza la hipótesis nula. Las calificaciones promedio de los usuarios para ambos géneros SON DIFERENTES.")
else:
    print("No se puede rechazar la hipótesis nula.")
```
