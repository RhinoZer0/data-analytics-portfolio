# Descripción del proyecto


Trabajas para la tienda online Ice que vende videojuegos por todo el mundo. Las reseñas de usuarios y expertos, los géneros, las plataformas (por ejemplo, Xbox o PlayStation) y los datos históricos sobre las ventas de juegos están disponibles en fuentes abiertas. Tienes que identificar patrones que determinen si un juego tiene éxito o no. Esto te permitirá detectar proyectos prometedores y planificar campañas publicitarias.

Delante de ti hay datos que se remontan a 2016. Imaginemos que es diciembre de 2016 y estás planeando una campaña para 2017.

Lo importante es adquirir experiencia de trabajo con datos. Realmente no importa si estás pronosticando las ventas de 2017 en función de los datos de 2016 o las ventas de 2027 en función de los datos de 2026.

El dataset contiene una columna "rating" que almacena la clasificación ESRB de cada juego. El Entertainment Software Rating Board (la Junta de clasificación de software de entretenimiento) evalúa el contenido de un juego y asigna una clasificación de edad como Adolescente o  proyecto

# TOP 6 PLATAFORMAS CON MÁS VENTAS

```python
#Creo una lista con las plataformas con más ventas
top_sales_platforms = ['PS2','X360','PS3','Wii','DS','PS']

#Creo un nuevo dataframe que incluya sólo los juegos cuya plataforma se incluya en la lista anterior
top_platforms = clean_games[clean_games['platform'].isin(top_sales_platforms)]

#Agrupo lo juegos por plataforma y año de lanzamiento
pivot_top_platforms = top_platforms.pivot_table(index=['platform','year_of_release'], values='total_sales', aggfunc='sum').reset_index()
# print(pivot_top_platforms)

#Agrupo nuevamente para cada plataforma, filtrando los juegos que no tenían año de lanzamiento
pivot_DS = pivot_top_platforms.query('platform=="DS" and year_of_release != 9999')[['year_of_release','total_sales']].set_index('year_of_release')
pivot_PS2 = pivot_top_platforms.query('platform=="PS2" and year_of_release != 9999')[['year_of_release','total_sales']].set_index('year_of_release')
pivot_X360 = pivot_top_platforms.query('platform=="X360" and year_of_release != 9999')[['year_of_release','total_sales']].set_index('year_of_release')
pivot_Wii = pivot_top_platforms.query('platform=="Wii" and year_of_release != 9999')[['year_of_release','total_sales']].set_index('year_of_release')
pivot_PS = pivot_top_platforms.query('platform=="PS" and year_of_release != 9999')[['year_of_release','total_sales']].set_index('year_of_release')
pivot_PS3 = pivot_top_platforms.query('platform=="PS3" and year_of_release != 9999')[['year_of_release','total_sales']].set_index('year_of_release')

#Creo los gráficos de baras para cada plataforma
fig, axes = plt.subplots(2,3, figsize=(25,10))
pivot_DS.plot(kind='bar',title='DS', xlabel='Year of release', ylabel='Total sales in million USD', ax= axes[0,0])
pivot_PS2.plot(kind='bar',title='PS2', xlabel='Year of release', ylabel='Total sales in million USD', ax= axes[0,1])
pivot_X360.plot(kind='bar',title='X360', xlabel='Year of release', ylabel='Total sales in million USD', ax= axes[0,2])
pivot_Wii.plot(kind='bar',title='Wii', xlabel='Year of release', ylabel='Total sales in million USD', ax= axes[1,0])
pivot_PS.plot(kind='bar',title='PS', xlabel='Year of release', ylabel='Total sales in million USD', ax= axes[1,1])
pivot_PS3.plot(kind='bar',title='PS3', xlabel='Year of release', ylabel='Total sales in million USD', ax= axes[1,2])
plt.tight_layout()
plt.show()
```

+ Las plataformas X360, PS y PS3 muestran una distribución normal de ventas, mientras que DS, PS2 y Wii muestran una asimetría positiva.

```python
#Creo una gráfica de barras que muestre todas las plataformas, los años en los que registran ventar y sus ventas totales en esos años
pivot_all_platforms=pd.merge(pivot_DS,pivot_PS,how='outer',on='year_of_release')
pivot_all_platforms.columns=['DS','PS']
pivot_all_platforms=pd.merge(pivot_all_platforms,pivot_PS2,how='outer',on='year_of_release')
pivot_all_platforms.columns=['DS','PS','PS2']
pivot_all_platforms=pd.merge(pivot_all_platforms,pivot_X360,how='outer',on='year_of_release')
pivot_all_platforms.columns=['DS','PS','PS2','X360']
pivot_all_platforms=pd.merge(pivot_all_platforms,pivot_Wii,how='outer',on='year_of_release')
pivot_all_platforms.columns=['DS','PS','PS2','X360','Wii']
pivot_all_platforms=pd.merge(pivot_all_platforms,pivot_PS3,how='outer',on='year_of_release')
pivot_all_platforms.columns=['DS','PS','PS2','X360','Wii','PS3']
pivot_all_platforms.plot(kind='bar',figsize=[25,10], title='Sales per year for all platforms', xlabel='Year of sales registered', ylabel='Total sales')
plt.show()
```

+ La plataforma PS muestra ventas a partir de 1994 y hasta 2003. Entre estos años aparece PS2 en el año 2000, 6 años después del inicio de ventas de PS. 9 años de ventas.
+ La plataforma PS2 muestra ventas a partir del 2000 y hasta 2011. Entre estos años aparece la plataforma DS en el 2004, 4 años después del inicio de ventas de PS2. 11 años de ventas.
+ La plataforma DS muestra ventas a partir del 2004 y hasta 2013. Entre estos años aparece la plataforma X360 en el 2005, sólo un año después del inicio de ventas de DS. 9 años de ventas
+ La plataforma X360 muestra ventas a partir del 2005 y hata 2016. Entre estos años aparece la plataforma Wii y PS3 en el 2006, sólo un año después del inicio de ventas de X360. 11 años de ventas.
+ Las plataformas Wii y PS3 muestran ventas a partir del 2006 hasta 2016. 10 años de ventas.
+ El tiempo de vida de las plataformas es de 10 años en promedio.
+ El tiempo que tarda una nueva plataforma en aparecer es de 5 años en promedio hasta antes de 2004, cuando este tiempo se reduce a sólo un año entre nuevas plataformas.

## ¿CÓMO AFECTAN LAS RESEÑAS A LAS VENTAS?

### Comparación de ventas totales de juegos en 2 de las plataformas más populares, PS2 y X360. Sólo para juegos con un score de la crítica mayor a 70 puntos y excluyendo todos los juegos que no tenían un score asignado (critic_score = 1000)

No parece haber alguna relación entre la plataforma y las ventas totales si comparamos estas dos plataformas. Algunos juegos tienen sólo la mitad de ventas en una plataforma que en la otra, aún tratándose del mismo juego.
En este análisis, el score de la crítica sólo se usó para filtrar los juegos y no presentar la lista completa.

```python
#Filtro los juegos por plataforma para PS2 y X360, tomando sólo los que tienen un score de la crítica payor a 70
PS2_games = clean_games[(clean_games['platform']=='PS2') & (clean_games['critic_score']>70) & (clean_games['critic_score']<1000)][['name','total_sales']]
X360_games = clean_games[(clean_games['platform']=='X360') & (clean_games['critic_score']>70) & (clean_games['critic_score']<1000)][['name','total_sales']]

#Mezclo las dos tablas para poder crear la gráfica de barras que muestre las ventas totales de juegos que se lanzaron en ambas plataformas
PS2_X360 = PS2_games.merge(X360_games, on='name', how='inner')
# display(PS2_X360)
PS2_X360.plot(kind='bar',figsize=[20,5],title='Games launched for PS2 and X360',x='name',xlabel='Game name',ylabel='Total sales')
plt.legend(['PS2','X360'])
plt.show()
```

## Calificaciones promedio de usuarios para plataformas Xbox One y PC

Hipótesis nula: Las calificaciones promedio de los usuarios para las plataformas Xbox One y PC son las mismas.

Hipótesis alternativa: Las calificaciones promedio de los usuarios para las plataformas Xbox One y PC SON DIFERENTES.

```python
#Filtro los juegos que ya tenían un score de usuario (distinto de 1000) y que pertenencen a las plataformas XOne y PC
xone_user_rating = clean_games.query('user_score<1000 and platform=="XOne"')
display(xone_user_rating)
pc_user_rating = clean_games.query('user_score<1000 and platform=="PC"')
display(pc_user_rating)
```

```python
# Prueba las hipótesis
alpha = 0.05
test = st.ttest_ind(xone_user_rating['user_score'],pc_user_rating['user_score'],False)
pvalue = test.pvalue 
print(pvalue)
if pvalue < alpha:
    print("Se rechaza la hipótesis nula. Las calificaciones promedio de los usuarios para ambas plataformas SON DIFERENTES.")
else:
    print("No se puede rechazar la hipótesis nula.")
```
